```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline
import math
```

```python
plt.figure()
x = np.arange(0, 120000, 1000)
x = list(map(lambda x: x/4380, x))
y = list(map(lambda x: math.exp((-1/8)*x), x))
sns.lineplot(x, y)
plt.yticks(np.arange(0,1.4, 0.2))
plt.xticks(np.arange(0,30,5))
plt.grid(visible=True)
plt.xlabel('Time (Years)')
plt.ylabel('Reliabiltiy')
plt.title('Reliability vs Time (Buck-Boost Converter, d>0.5)')
plt.xlim(left=0)
plt.show()
```
